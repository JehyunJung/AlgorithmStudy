```python
import numpy as np
import matplotlib.pyplot as plt
import time
```

```python
def sequentialSearch(data,value):
    for idx,item in enumerate(data):
        if item==value:
                return idx
    else :
        return -1
```

```python
def binarySearch(data,left,right,value):
    if len(data)<=0:
        return -1
    medium=(left+right)//2
    if data[medium]==value:
        return medium
    elif data[medium]<value:
        return binarySearch(data,medium+1,right,value)
    else:
        return binarySearch(data,left,medium-1,value)
```

```python
def binarySearch_iteration(data,value):
    left,right=0,len(data)
    while left<=right:
        med=(left+right)//2
        if data[med]==value:
            return med
        elif data[med]<value:
            left=med+1
        else:
            right=med-1
    return -1
```

```python
X_range=[5,10,100,500,1000,2000]
def TimeComplexity(X_range,Searching_Method):
    times=[]
    for i in X_range:
        np.random.seed(i)
        data=np.random.randint(0,1000,size=i)
        start=time.time()
        if Searching_Method == None:
            np.random.seed(i)
            value=np.random.randint(0,1000,size=1)
            data.sort()
            list(data).index(value)
        elif Searching_Method == binarySearch:
            np.random.seed(i)
            value=np.random.randint(0,1000,size=1)
            data.sort()
            data=Searching_Method(data,0,len(data),value)
        else:
            np.random.seed(i)
            value=np.random.randint(0,1000,size=1)
            data.sort()
            data=Searching_Method(data,value)
        times.append(time.time()-start)
    return times

plt.plot(X_range,TimeComplexity(X_range,sequentialSearch),label="Sequential Search")
plt.plot(X_range,TimeComplexity(X_range,binarySearch),label="Binary Search(rec)")
plt.plot(X_range,TimeComplexity(X_range,binarySearch_iteration),label="Binary Search(itr)")
plt.plot(X_range,TimeComplexity(X_range,None),label="Builtin_Search")
plt.title('Data Searching')
plt.xlabel('data count')
plt.ylabel('time(second)')
plt.legend()
plt.show()
```
